```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("../data/processed/cleaned_medical_appointments.csv")
df.head()
```

```python
df.shape
```

```python
df['no_show'].value_counts()
```

```python
df['no_show'].value_counts(normalize=True) * 100
```

```python
df['no_show'].value_counts().plot(kind='bar')
plt.title('No-show Distribution')
plt.xlabel('No-show (0 = Attended, 1 = Missed)')
plt.ylabel('Number of Patients')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='age', hue='no_show', bins=30, kde=True)

plt.title('Age Distribution by No-show')
plt.xlabel('Age')
plt.ylabel('Count')

plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='sms_received', hue='no_show')

plt.title('SMS Received vs No-show')
plt.xlabel('SMS Received (0 = No, 1 = Yes)')
plt.ylabel('Number of Patients')

plt.show()
```

```python
pd.crosstab(
    df['sms_received'],
    df['no_show'],
    normalize='index'
) * 100
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(data=df, x='no_show', y='waiting_days')

plt.title('Waiting Days vs No-show')
plt.xlabel('No-show (0 = Attended, 1 = Missed)')
plt.ylabel('Waiting Days')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(data=df, x='no_show', y='age')

plt.title('Age vs No-show')
plt.xlabel('No-show (0 = Attended, 1 = Missed)')
plt.ylabel('Age')

plt.show()
```
