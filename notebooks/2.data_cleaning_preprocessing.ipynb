```python
import pandas as pd
import numpy as np
df=pd.read_csv("../data/dataset.csv")
df.head()
```

```python
df.columns
```

```python
df.rename(columns={
    'PatientId': 'patient_id',
    'AppointmentID': 'appointment_id',
    'Gender': 'gender',
    'ScheduledDay': 'scheduled_day',
    'AppointmentDay': 'appointment_day',
    'Age': 'age',
    'Neighbourhood': 'neighbourhood',
    'Scholarship': 'scholarship',
    'Hipertension': 'hypertension',
    'Diabetes': 'diabetes',
    'Alcoholism': 'alcoholism',
    'Handcap': 'handicap',
    'SMS_received': 'sms_received',
    'No-show': 'no_show'
}, inplace=True)
```

```python
df.columns
```

```python
df.dtypes
```

```python
df['scheduled_day'] = pd.to_datetime(df['scheduled_day'])
df['appointment_day'] = pd.to_datetime(df['appointment_day'])
```

```python
df['no_show'] = df['no_show'].map({
    'No': 0,
    'Yes': 1
})
```

```python
df['no_show'].value_counts()
```

```python
df['waiting_days'] = (df['appointment_day'] - df['scheduled_day']).dt.days
```

```python
df[['scheduled_day', 'appointment_day', 'waiting_days']].head()
```

```python
df['appointment_weekday'] = df['appointment_day'].dt.day_name()
```

```python
df[['appointment_day', 'appointment_weekday']].head()
```

```python
df['scheduled_hour'] = df['scheduled_day'].dt.hour
df[['scheduled_day', 'scheduled_hour']].head()
```

```python
df[df['waiting_days'] < 0][['scheduled_day', 'appointment_day', 'waiting_days']].head()
```

```python
df['scheduled_day'] = df['scheduled_day'].dt.normalize()

df['waiting_days'] = (df['appointment_day'] - df['scheduled_day']).dt.days
```

```python
df['waiting_days'].describe()
```

```python
(df['waiting_days'] < 0).sum()
```

```python
df[df['waiting_days'] < 0][['scheduled_day', 'appointment_day', 'waiting_days']]
```

```python
df = df[df['waiting_days'] >= 0]
```

```python
df.drop(columns=[
    'patient_id',
    'appointment_id',
    'scheduled_day',
    'appointment_day'
], inplace=True)
```
