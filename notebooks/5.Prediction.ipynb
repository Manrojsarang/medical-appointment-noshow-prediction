```python
import pandas as pd
import joblib

```

```python
model = joblib.load("../models/medical_no_show_model.pkl")
```

```python
feature_columns = joblib.load("../models/feature_columns.pkl")
```

```python
new_patient = pd.DataFrame(
    0,
    index=[0],
    columns=feature_columns
)
```

```python
new_patient.head()
```

```python
new_patient["gender"] = 0
new_patient["age"] = 25
new_patient["scholarship"] = 0
new_patient["hypertension"] = 0
new_patient["diabetes"] = 0
new_patient["alcoholism"] = 0
new_patient["sms_received"] = 1
new_patient["waiting_days"] = 30
new_patient["scheduled_hour"] = 10

new_patient["appointment_weekday_Monday"] = 1
new_patient["neighbourhood_CENTRO"] = 1
```

```python
prediction = model.predict(new_patient)

if prediction[0] == 1:
    print("Prediction: Patient is likely to MISS the appointment.")
else:
    print("Prediction: Patient is likely to ATTEND the appointment.")


    
```
